```python
#!/usr/bin/env python
# coding: utf-8

import os
import requests
from bs4 import BeautifulSoup

cnt = 0
picList=[]
for i in range(1):
    link='https://photo.xuite.net/starandsea/20431003*%s'%i
    res = requests.get(link)
    soup = BeautifulSoup(res.text, "html.parser")
    print('page%s'%i)
    for data in soup.select('.photo_cover'):
        picLink = data['src']
        picLink = picLink.split('_')
        picLink = '%s%s'%(picLink[0],'_x.jpg')
        print(cnt)
        cnt+=1
#         print(picLink)
        if picLink not in picList:
            picList.append(picLink)
            
cnt = 0
for link in picList:
    cnt+=1
    print(link)
    #os.system('wget %s'%link)
print(cnt)
print("=======Done========")
```
